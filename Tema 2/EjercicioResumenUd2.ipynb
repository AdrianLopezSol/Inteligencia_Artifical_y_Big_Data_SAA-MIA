```python
import pandas as pd
import numpy as np

datos_ventas = {
    'Fecha': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-08', '2023-01-08', '2023-01-09'],
    'Tienda_ID': ['T1', 'T2', 'T1', 'T2', 'T1', 'T2', 'T1', 'T2'],
    'Producto': ['Café', 'Tarta', 'Café', 'Pastel', 'Té', 'Café', 'Tarta', 'Pastel'],
    'Cantidad': [20, 5, 25, 7, 10, 18, 6, 8]
}

datos_tiendas = {
    'Tienda_ID': ['T1', 'T2'],
    'Ciudad': ['Valencia', 'Madrid']
}
```

Hacemos los dataframes sobre los que trabajaremos

```python
df_ventas = pd.DataFrame(datos_ventas)
df_tiendas = pd.DataFrame(datos_tiendas)
```

¿Cuál es la cantidad total de productos vendidos en cada ciudad?

```python
df_completo = pd.merge(df_ventas, df_tiendas, on='Tienda_ID', how='outer')
total_por_ciudad = df_completo.groupby('Ciudad')['Cantidad'].sum().reset_index()
print (total_por_ciudad)
```

¿Cómo se distribuyen las ventas de cada producto por ciudad?

```python
df_producto_y_ciudad = df_completo.pivot_table(values='Cantidad', index='Ciudad', columns='Producto')
print(df_producto_y_ciudad.fillna(0))
```

¿Cuál ha sido el total de productos vendidos por semana?

```python
ventas_semanales = df_completo.resample('W', on='Fecha')['Cantidad'].sum().reset_index()
print (ventas_semanales)
```
